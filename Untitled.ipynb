```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
df.tail()
```

```python
Adelie_df = df.loc[df['species'] == 'Adelie']
Gentoo_df = df.loc[df['species'] == 'Gentoo']
my_list = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
for i in my_list:
    print(i)
    my_list2 = my_list.copy()
    my_list2.remove(i)
    for j in my_list2:
        Adelie_1 = Adelie_df[i].tolist()
        Gentoo_1 = Gentoo_df[i].tolist()
        Adelie_2 = Adelie_df[j].tolist()
        Gentoo_2 = Gentoo_df[j].tolist()
        plt.figure()  
        plt.scatter(Adelie_1, Adelie_2)
        plt.scatter(Gentoo_1, Gentoo_2)
        plt.xlabel(i)
        plt.ylabel(j)
```

```python
cond = df['species'].isin(['Adelie','Gentoo'])
penguin_data = df.loc[cond,['bill_depth_mm','flipper_length_mm']]
penguin_data.head()
```

```python
penguin_target = df.loc[cond,['species']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(penguin_data, penguin_target)
```

```python
train_input
```

```python
flipper_length_mm_mean = df['flipper_length_mm'].mean()
df['flipper_length_mm'] = df['flipper_length_mm'].fillna(flipper_length_mm_mean)
```

```python
bill_depth_mm_mean = df['bill_depth_mm'].mean()
df['bill_depth_mm'] = df['bill_depth_mm'].fillna(bill_depth_mm_mean)
```

```python
df.isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target['species'])
```

```python
kn.score(test_input, test_target['species'])
```

```python
from sklearn.preprocessing import LabelEncoder

# 범주형 타겟 변수 숫자로 변환
le = LabelEncoder()
df['species_encoded'] = le.fit_transform(df['species'])

# 독립변수(X)와 변환된 타겟변수(y) 간의 상관계수 계산
X = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = df['species_encoded']

corr_matrix = X.corrwith(y)
best_features = corr_matrix.abs().nlargest(2).index.tolist()
print("선택된 독립변수:", best_features)
```
